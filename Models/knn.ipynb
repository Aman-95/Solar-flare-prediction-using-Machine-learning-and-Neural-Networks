#**K-Nearest Neighbour**

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
dataset = pd.read_excel("Dataset_set_2_for_classification.xlsx")
```

```
dataset
```

# For c-class solar flares

```
X = dataset.iloc[1:, 0:38].values
y = dataset.iloc[1:, 46].values
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 0)
```

```
from sklearn.preprocessing import StandardScaler
sc_X_train = StandardScaler()
sc_y_train = StandardScaler()
X_train = sc_X_train.fit_transform(X_train)
X_test = sc_X_train.transform(X_test)
#y_train = sc_y_train.fit_transform(y_train)
#y_test = sc_y_train.transform(X_test)
```

```
y_train=np.asarray(y_train,dtype=float)
```

```
# Fitting K-NN to the Training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```
y_pred = classifier.predict(X_test)
y_pred = np.asarray(y_pred,dtype=int)
```

```
y_test=np.asarray(y_test,dtype=int)
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
